```python
# Step 1: Install required library
!pip install PyPDF2

# Step 2: Import
import PyPDF2
from google.colab import files

# Step 3: Upload PDFs
print("📤 Upload the PDF files you want to merge:")
uploaded = files.upload()
all_files = list(uploaded.keys())

# Step 4: Show uploaded files
print("\n📄 Uploaded files:")
for i, fname in enumerate(all_files):
    print(f"{i+1}. {fname}")

# Step 5: Ask for correct order
print("\n✏️ Enter the filenames to merge in correct order, separated by commas.")
print("Example: 1.pdf, 3.pdf, 2.pdf")
order_input = input("📝 Your order: ")

# Convert input into list
input_pdfs = [x.strip() for x in order_input.split(',')]

# Step 6: Merge function
def merge_pdfs(input_pdfs, output_pdf):
    pdf_writer = PyPDF2.PdfWriter()
    for input_pdf in input_pdfs:
        with open(input_pdf, 'rb') as f:
            reader = PyPDF2.PdfReader(f)
            for page in reader.pages:
                pdf_writer.add_page(page)
    with open(output_pdf, 'wb') as out:
        pdf_writer.write(out)

# Step 7: Merge and auto-download
output_pdf = "merged_output.pdf"
merge_pdfs(input_pdfs, output_pdf)
print("✅ Merging complete. Downloading...")
files.download(output_pdf)
```
